```python
import numpy as np

rng = np.random.default_rng(seed=1)
```

```python
population = np.array([0, 1, 2])
print(f"mean: {round(population.mean(), 3)}")
print(f"var: {round(population.var(ddof=0), 3)}")

mean_list, var_list = [], []
for _ in range(10_000):
    sample = rng.choice(population, 2, replace=True)
    mean_list.append(np.mean(sample))
    var_list.append(np.var(sample, ddof=1))

print(f"mean sample mean: {round(np.mean(mean_list), 3)}")
print(f"mean sample var: {round(np.mean(var_list), 3)}")
```

```python
import seaborn as sns
from scipy.stats import norm, t

samples_means = []

for _ in range(10_000):
    sample = rng.normal(0, 1, 4)
    samples_means.append(np.sqrt(4) * np.mean(sample) / np.std(sample, ddof=1))

sns.kdeplot(samples_means, clip=[-3, 3], label='sample')
x = np.arange(-3, 3, 0.01)
sns.lineplot(x=x, y=norm.pdf(x), label='z')
sns.lineplot(x=x, y=t.pdf(x, 3), label='t')
```
