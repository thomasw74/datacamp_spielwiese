```python
import requests

response = requests.get('https://de.wikipedia.org/wiki/Liste_der_Staaten_der_Erde')
response.status_code, response.encoding
```

```python
from scrapy import Selector

selector = Selector(text=response.text)
tables = selector.xpath('//table')
tables
```

```python
table_selector = tables[0]
table_selector.extract()[:100]
```

```python
header_selectors = table_selector.css('tr:nth-of-type(1) > th')
for h in header_selectors:
    print(" ".join(h.css(' ::text').extract()[:-1]))
```

```python
with open('data/ship_table.txt') as file:
    ship_table = file.read()
ship_selector = Selector(text=ship_table)
```

```python
ships = ship_selector.css('a::text').extract()
ships[:5]
```

```python
from collections import Counter

def optionals_todict(s):
    return dict(Counter(map(int, s.split())))

optionals = list(map(optionals_todict, ship_selector.css('tr > td:last-child::text').extract()))
optionals[:5]
```
